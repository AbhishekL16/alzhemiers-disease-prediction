```python
import cv2
import keras
from tensorflow.keras.models import load_model
from keras.models import model_from_json
from pathlib import Path
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class_names = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']
```

```python
model = load_model("model/cnn_model.h5")
```

```python
import os
user_image = os.listdir("u_input")
```

```python
user_image
```

```python
user_input = int(input("enter the file index number:"))
```

```python
user_select_image = user_image[user_input]
```

```python
user_select_image
```

```python
path=os.path.join("u_input",user_select_image)
```

```python
print(path)
```

```python
img = cv2.imread(path, 1)
```

```python
plt.imshow(img)
```

```python
img = np.array(cv2.resize(img, (128, 128)))
img = img/255
```

```python
img = img.reshape(1,img.shape[0],img.shape[1],img.shape[2])
results = model.predict(img)
```

```python
results
```

```python
single_result = results[0]
```

```python
most_likely_class_index = int(np.argmax(single_result))
class_likelihood = single_result[most_likely_class_index]  
```

```python
most_likely_class_index
```

```python
class_label = class_names[most_likely_class_index]
```

```python
print("Predicted class is  {} - Confidence Score: {:2f}".format(class_label, class_likelihood))
```
